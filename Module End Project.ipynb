```python
# Loading dataset
data = pd.read_csv("C:\\Users\\Faris\\Jupyter\\myexcel - myexcel.csv.csv")
import numpy as np
random_heights= np.random.randint(150,181,size=len(data))
data['Height']=random_heights
data
```

```python
Team_group = data.groupby('Team')
for team_name, team_data in Team_group:
    print(f"Team: {team_name}")
    team_report = team_data.describe()
    print(team_report)
    print("\n\n")
```

```python
#Grouping the dataframe by 'Team' and counting the number of employees
Team_count=data['Number'].groupby(data['Team']).count()
Team_count
```

```python
#Calculating the percentage split
total_employees = data['Number'].count()
percentage_split = (Team_count/total_employees)*100
percentage_split = percentage_split.round(3)
```

```python
#combining counts and percentage split into a dataframe for easy comparison
team_summery = pd.DataFrame({'Number of Employees':Team_count,'Percentage split (%)':percentage_split})
team_summery
```

```python
#grouping employees by their positions
position_group = data.groupby('Position')
for position, employees in position_group:
    print(f"Employees in position : {position}")
    print(employees)
    print("\n")
```

```python
#defining age groups
bins = [18,22,27,32,37,42,47]
labels = ['18-21','22-26','27-31','32-36','37-41','42+']
data['Age_group'] = pd.cut(data['Age'],bins=bins,labels=labels,right=False)
```

```python
#finding the age group most number of employees belongs to
most_common_age_group=data['Number'].groupby(data['Age_group']).count()
most_common_age_group
```

```python
#Finding the sum of salary regards to team and position
Team_Position_Salary= data.groupby(['Team','Position'])['Salary'].sum()
```

```python
#finding the maximum spending in correspondence to Team and Position
max_spending=Team_Position_Salary.idxmax()
max_spending_amount = Team_Position_Salary.max()
```

```python
print(f"Highest Spending interms of  salary is in :\nTeam: {max_spending[0]}, Position:{max_spending[1]}")
print(f"Total Spending in this Team:{max_spending_amount}")
```

```python
#plotting scatterplot
plt.figure(figsize=(8,6))
plt.scatter(data['Age'],data['Salary'], alpha=0.5)
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
Yvalue = np.linspace(50000,25000000,num=5)
plt.yticks(Yvalue)
plt.show()
```
